Yapılan uzun araştırmalar sonucu, yaşın artışıyla kçük değişmeler olmakla beraber sistolik tansiyon (büyük) için 120 ile 140, diastolik (küçük) tansiyon için 70-90 arası olması halinde tansiyona bağlı olarak bir sağlık sorunu riski doğmadığı belirlenmiştir. Klavyeden girilen sıfırdan farklı büyük ve küçük tansiyon değerine göre sağlık riski olup olmadığını ekrana yazmaya devam eden fonksiyonu yazınız.

WhatsApp Image 2022-04-11 at 17.01.29.jpeg

```python
def tansiyonKontrol(tansiyon,altSinir,ustSinir,dusuk,normal,yuksek):
  if tansiyon>=altSinir and tansiyon<=ustSinir:
    normal+=1
  elif tansiyon<altSinir:
    dusuk+=1
  else:
    yuksek+=1
  return dusuk,normal,yuksek
```

```python
buyukTansiyon=int(input('Büyük tansiyon giriniz: '))
kucukTansiyon=int(input('Küçük tansiyon giriniz: '))
bNormal=0
bYuksek=0
bDusuk=0
kNormal=0
kDusuk=0
kYuksek=0
while buyukTansiyon!=0 and kucukTansiyon!=0:

  bDusuk,bNormal,bYuksek=tansiyonKontrol(buyukTansiyon,120,140,bDusuk,bNormal,bYuksek)
  kDusuk,kNormal,kYuksek=tansiyonKontrol(kucukTansiyon,70,90,kDusuk,kNormal,kYuksek)
 
  buyukTansiyon=int(input('Buyuk tansiyon giriniz: '))
  kucukTansiyon=int(input('Kucuk tansiyon giriniz: '))
print('Büyük (Normal=',bNormal,'Yüksek=',bYuksek,'Düşük=',bDusuk,')')
print('Küçük (Normal=',kNormal,'Yüksek=',kYuksek,'Düşük=',kDusuk,')')
```
