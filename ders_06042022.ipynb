Lisansüstü eğitim almak için başvuran adayın ALES puanı, yabancı dil puanı ve mezuniyet not ortalamasına göre bir sıralama puanı hesaplanır. Yüksek lisans için başvuru yapıyor ve sıralama puanı eğer 60'ın üstünde ise veya Doktora için başvuru yapıyor ve sıralama puanı 80'in üzerinde ise sıralamaya girebilir. Sıralama puanı hesaplaması Ales %50, yabancı dil %25 ortalama %25 şeklinde yapılacaktır. Hesaplama sonucu Yds en az yüksek lisans için 50, doktora için 70 olmalıdır. Doktora başvuru için yüksek lisans notu 75'den büyük olmalıdır. Bunun için bir kod tasarlayınız

```python
ales=float(input('ales puaninizi giriniz: '))
yds=float(input('yds puaninizi giriniz: '))
notOrt=float(input('not ortalamanizi puaninizi giriniz: '))
hesaplama= (ales*0.5)+(yds*0.25)+(notOrt*0.25)
secim=input('yuksek lisans=1 doktora=2: ')

if secim=='2':
  if yds>=70:
    yuksekLisansOrt=float(input('yuksek lisans ort giriniz: '))
    if yuksekLisansOrt>=75:
      if hesaplama>=80:
        print('doktoraya basvurabilirsiniz')
      else:
        print('doktoraya basvuramazsiniz')
    else:
      print('yuksek lisans ortalamaniz yetersiz')
  else:
    print('yds puaniniz yetersiz')

elif secim=='1':
  if yds>=50:
    if hesaplama>=60:
      print('yuksek lisansa basvurabilirsiniz')
    else:
      print('yuksek lisansa basvuramazsiniz')
  else:
    print('yds puaniniz yetersiz') 
 
else:
  print('secim icin 1 veya 2 ye basiniz')
```

Ekrandan istenen kalma süresine göre otopark ücretini hesaplayıp ekrana yazdıran kodu tasarlayınız.

+ Süre < 1 saat ise ücret:5TL
+ 1<= süre < 5 saat ise ücret:10TL
+ 5<= süre < 10 saat ise ücret:17TL
+ süre >=10 ise 10. saatten sonraki her saat için 5TL

```python
sure=float(input('Kaldiginiz sureyi saat cinsinden giriniz: '))
if sure<1:
  print('ucret: 5 TL')
elif sure>=1 and sure<5:
  print('ucret: 10 TL')
elif sure>=5 and sure<10:
  print('ucret: 17 TL')
elif sure>=10:
  hesap=(int(sure)-10)*5
  hesap+=17
  print('ucret: '+str(hesap)+' TL')
else:
  print('gecerli bir sure giriniz')
```

Klavyeden girilen girilen geometrik şekil bilgisine göre şeklin alanını hesaplayan kodu yazınız. Alanı hesaplayabilmek için yine kullanıcıdan kenar uzunluğu, yükseklik gibi veriler istenecektir. Daire, Dikdörtgen, Kare, Üçgen (Alan hesaplamak için ekrandan farklı değerler istemek gerekebilir)

```python
sekilBilgisi=input('Hangi geometrik seklin alaninin hesaplamak istiyorunuz (1:daire 2:dikdortgen 3:kare 4:ucgen): ')
if sekilBilgisi=='1':
  yaricap=float(input('yaricapi giriniz: '))
  hesap=3.14*yaricap*yaricap
elif sekilBilgisi=='2':
  uzunKenar=float(input('uzun kenar olcusu: '))
  kisaKenar=float(input('kisa kenar olcusu: '))
  hesap=uzunKenar*kisaKenar
elif sekilBilgisi=='3':
  kenar=float(input('kenar olcusunu giriniz: '))
  hesap=kenar*kenar
elif sekilBilgisi=='4':
  yukseklik=float(input('yukseklik olcusu: '))
  tabanKenar=float(input('taban kenar olcusu: '))
  hesap=(yukseklik*tabanKenar)/2
else:
  print('gecersiz giris')
print('alan: '+ str(hesap))
```
