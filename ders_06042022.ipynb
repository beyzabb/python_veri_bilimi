Lisansst eitim almak iin bavuran adayn ALES puan, yabanc dil puan ve mezuniyet not ortalamasna gre bir sralama puan hesaplanr. Yksek lisans iin bavuru yapyor ve sralama puan eer 60'n stnde ise veya Doktora iin bavuru yapyor ve sralama puan 80'in zerinde ise sralamaya girebilir. Sralama puan hesaplamas Ales %50, yabanc dil %25 ortalama %25 eklinde yaplacaktr. Hesaplama sonucu Yds en az yksek lisans iin 50, doktora iin 70 olmaldr. Doktora bavuru iin yksek lisans notu 75'den byk olmaldr. Bunun iin bir kod tasarlaynz

```python
ales=float(input('ales puaninizi giriniz: '))
yds=float(input('yds puaninizi giriniz: '))
notOrt=float(input('not ortalamanizi puaninizi giriniz: '))
hesaplama= (ales*0.5)+(yds*0.25)+(notOrt*0.25)
secim=input('yuksek lisans=1 doktora=2: ')

if secim=='2':
  if yds>=70:
    yuksekLisansOrt=float(input('yuksek lisans ort giriniz: '))
    if yuksekLisansOrt>=75:
      if hesaplama>=80:
        print('doktoraya basvurabilirsiniz')
      else:
        print('doktoraya basvuramazsiniz')
    else:
      print('yuksek lisans ortalamaniz yetersiz')
  else:
    print('yds puaniniz yetersiz')

elif secim=='1':
  if yds>=50:
    if hesaplama>=60:
      print('yuksek lisansa basvurabilirsiniz')
    else:
      print('yuksek lisansa basvuramazsiniz')
  else:
    print('yds puaniniz yetersiz') 
 
else:
  print('secim icin 1 veya 2 ye basiniz')
```

Ekrandan istenen kalma sresine gre otopark cretini hesaplayp ekrana yazdran kodu tasarlaynz.

+ Sre < 1 saat ise cret:5TL
+ 1<= sre < 5 saat ise cret:10TL
+ 5<= sre < 10 saat ise cret:17TL
+ sre >=10 ise 10. saatten sonraki her saat iin 5TL

```python
sure=float(input('Kaldiginiz sureyi saat cinsinden giriniz: '))
if sure<1:
  print('ucret: 5 TL')
elif sure>=1 and sure<5:
  print('ucret: 10 TL')
elif sure>=5 and sure<10:
  print('ucret: 17 TL')
elif sure>=10:
  hesap=(int(sure)-10)*5
  hesap+=17
  print('ucret: '+str(hesap)+' TL')
else:
  print('gecerli bir sure giriniz')
```
